```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Marvel_DC_imdb.csv")
print(len(df.index))
df.head()
```

```python
df = df.dropna(axis=0, subset=['RunTime','IMDB_Score','USA_Gross'])
df.info()
```

```python
df['RunTime'] = df['RunTime'].map(lambda x: x.rstrip(' min'))
df["RunTime"] = pd.to_numeric(df["RunTime"], downcast="integer")
df = df.sort_values('RunTime',ascending=False)
df = df.dropna(axis=0, subset=['RunTime'])
```
